# Netflix Data Cleaning, Analysis and VisualizationThis notebook performs data cleaning, exploratory data analysis (EDA), and visualization on the Netflix dataset.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

## 2. Load Dataset

```python
data = pd.read_csv('netflix1.csv')
data.head()
```

## 3. Dataset Overview

```python
data.shape
data.columns
data.info()
```

## 4. Check Missing Values

```python
data.isnull().sum()
```

## 5. Remove Duplicate Records

```python
data = data.drop_duplicates()
data.shape
```

## 6. Convert Date Column

```python
data['date_added'] = pd.to_datetime(data['date_added'])
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month
```

## 7. Content Type Distribution

```python
type_counts = data['type'].value_counts()
type_counts.plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.show()
```

## 8. Content Rating Distribution

```python
rating_counts = data['rating'].value_counts()
rating_counts.plot(kind='bar')
plt.title('Content Rating Distribution')
plt.show()
```

## 9. Top Countries Producing Content

```python
top_countries = data['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title('Top Countries on Netflix')
plt.show()
```

## 10. Content Added Over Time

```python
yearly_content = data['year_added'].value_counts().sort_index()
yearly_content.plot(kind='line')
plt.title('Content Added Over Years')
plt.show()
```

## 11. Most Popular Genres

```python
genre_counts = data['listed_in'].value_counts().head(10)
genre_counts.plot(kind='bar')
plt.title('Top Genres on Netflix')
plt.show()
```

## 12. Top Directors

```python
top_directors = data['director'].value_counts().head(10)
top_directors.plot(kind='bar')
plt.title('Top Directors on Netflix')
plt.show()
```

## 13. Word Cloud of Titles

```python
text = ' '.join(data['title'].astype(str))
wordcloud = WordCloud(width=800, height=400).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## 14. Key Insights• Movies dominate Netflix compared to TV shows.• The United States contributes the largest number of titles.• Netflix content increased rapidly after 2016.• Drama and International genres are very common.
